```python
import win32com.client as win32

class execute:
    
    def __init__(self, MainFolder = 'Inbox', SubFolder = 'Kroger DSDs'):

        self.MainFolder = MainFolder
        self.SubFolder = SubFolder
        
        self.getFolders()
        self.getMessages()
    
    def getFolders(self):
        
        self.outlook = win32.Dispatch('Outlook.Application').GetNamespace('MAPI')

        i = 0
        
        self.inbox = None

        while str(self.inbox) != str(self.MainFolder):
            try:
                self.inbox = self.outlook.GetDefaultFolder(i)
            except:
                pass
            i = i+1

            if i > 300:

                raise ValueError(f'Cannot find folder {self.MainFolder}')


        if not self.SubFolder == None:

            i = 0
            self.folder = None

            while str(self.folder) != str(self.SubFolder):

                try:
                    self.folder = self.inbox.Folders(i)
                except:
                    pass
                i = i+1
                
                if i > 300:
                    
                    raise ValueError(f'Cannot find folder {self.SubFolder}')


        else:
            self.folder = self.inbox
            
        self.messages = self.folder.Items
        
    def getMessages(self):
        
        import re
        import os
      
        
        for message in self.messages:
            if message.Unread == True:
                for attachment,i in zip(message.Attachments,range(len(message.Attachments))):
                    if '.png' not in str(attachment):
                        aname = re.sub(' ','',str(attachment))+str(i)
                        file = r"R:\AR\Audit\Kroger\2021\FPP/"+aname+'.msg'
                        try:
                            attachment.SaveAsFile(file)
                        except:
                            pass
                        self.getExcelData(file,i)
                        os.remove(file)
                        
                        message.Unread = False
                        

    def getExcelData(self, file,i):
        
        import re
        import os

        msg = self.outlook.OpenSharedItem(file)
        for attachment in msg.Attachments:
            root = r"<<FILEPATH>>"
            print(i)
            x_file = root+'/'+str(attachment).split('.')[0] + str(i) + '.' + str(attachment).split('.')[1]
            attachment.SaveAsFile(x_file)
            
            
            
            
```
